# Generación de graficos.

```python
# Leer el archivo CSV
file_path = 'primos_tiempos.csv'
df = pd.read_csv(file_path)

# Reemplazar los valores de "Primo en posición"
valores_reemplazo = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
df['Primo en posición'] = valores_reemplazo

# Configuración del estilo de la gráfica para formato IEEE
sns.set_context("paper", font_scale=1.2)
sns.set_style("whitegrid")

# Crear la gráfica
plt.figure(figsize=(5, 4))
sns.lineplot(
    data=df,
    x='Primo en posición',
    y='Tiempo de ejecución (s)',
    marker='o',
    color="royalblue",
    linewidth=2.5
)

# Personalizar etiquetas y título
plt.xlabel("Primos calculados (n)", fontsize=6, labelpad=7)
plt.ylabel("Tiempo de ejecución (segundos)", fontsize=6, labelpad=7)
plt.title("Relación entre tiempo de ejecución y cantidad de primos", fontsize=12, pad=20)

# Ajustar escala logarítmica en X
plt.xscale('log')

# Limpiar bordes y mejorar visibilidad
sns.despine(trim=True)

# Guardar o mostrar la gráfica
plt.tight_layout()
plt.savefig("grafica_ieee_pulcro.png", dpi=400, bbox_inches='tight')
plt.show()
```

```python
# Cargar los datos
file_path = '30_runs_normal_mint.csv'
data = pd.read_csv(file_path)

# Transformar los datos para comprimir el rango del eje X
def ajustar_eje_x(valor):
    if valor >= 80:
        return valor - 15
    return valor

data['Tiempo ajustado (s)'] = data['Tiempo de ejecución (s)'].apply(ajustar_eje_x)

# Configuración del estilo para gráficos estilo paper
sns.set_context("paper", font_scale=1.2)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6, 4))

# Crear el histograma ajustado con Seaborn
sns.histplot(
    data['Tiempo ajustado (s)'],
    bins=30,
    kde=True,
    color='royalblue',
    edgecolor='black',
    alpha=0.7,
)

# Personalizar las etiquetas del eje X
ticks_originales = list(range(40, 86, 5))  # Rangos originales
ticks_ajustados = [ajustar_eje_x(tick) for tick in ticks_originales]
etiquetas = [str(tick) if tick < 65 or tick >= 80 else '' for tick in ticks_originales]
plt.xticks(ticks_ajustados, etiquetas, fontsize=10)

# Personalización del gráfico
plt.title('Histograma del Tiempo de Ejecución', fontsize=12, weight='bold')
plt.xlabel('Tiempo de Ejecución (s)', fontsize=10)
plt.ylabel('Frecuencia', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Ajustar diseño y guardar el gráfico como PDF
plt.tight_layout()
plt.savefig("histograma_tiempo_ejecucion.pdf", bbox_inches='tight', format='pdf')
plt.show()

```

# Test estadistico $t$ para encontrar si existen tiempos significativamente menores.

```python
# Nivel de significancia
alpha = 0.05

# Función para realizar un one-tailed t-test
def one_tailed_ttest(sample1, sample2, alpha):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)
    # Ajustar el test para una cola
    if t_stat < 0:  # La media de sample1 (parallel) es menor que la de sample2 (normal)
        one_tailed_p = p_value / 2
        if one_tailed_p < alpha:
            return True, one_tailed_p  # Diferencia significativa en una cola
        else:
            return False, one_tailed_p  # No hay diferencia significativa
    else:
        return False, 1  # Dirección opuesta, no significativa

# Test para Mint
mint_result, mint_p_value = one_tailed_ttest(data_mint_parallel, data_mint_normal, alpha)
if mint_result:
    print(f"Para Mint, los tiempos paralelizados son significativamente menores (one-tailed, p-value: {mint_p_value:.4f})")
else:
    print(f"Para Mint, NO se encontró una diferencia significativa en la dirección esperada (p-value: {mint_p_value:.4f})")

# Test para Ubuntu
ubuntu_result, ubuntu_p_value = one_tailed_ttest(data_ubuntu_parallel, data_ubuntu_normal, alpha)
if ubuntu_result:
    print(f"Para Ubuntu, los tiempos paralelizados son significativamente menores (one-tailed, p-value: {ubuntu_p_value:.4f})")
else:
    print(f"Para Ubuntu, NO se encontró una diferencia significativa en la dirección esperada (p-value: {ubuntu_p_value:.4f})")
```

```python
# Nivel de significancia
alpha = 0.05

# Función para realizar un one-tailed t-test
def one_tailed_ttest(sample1, sample2, alpha):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)
    # Ajustar el test para una cola
    if t_stat < 0:  # La media de sample1 es menor que la de sample2
        one_tailed_p = p_value / 2
        if one_tailed_p < alpha:
            return True, one_tailed_p  # Diferencia significativa en una cola
        else:
            return False, one_tailed_p  # No hay diferencia significativa
    else:
        return False, 1  # Dirección opuesta, no significativa

# Test para la categoría "Normal"
normal_result, normal_p_value = one_tailed_ttest(data_mint_normal, data_ubuntu_normal, alpha)
if normal_result:
    print(f"En la categoría 'Normal', Linux Mint tiene tiempos significativamente menores que Ubuntu (one-tailed, p-value: {normal_p_value:.4f})")
else:
    print(f"En la categoría 'Normal', NO se encontró una diferencia significativa en la dirección esperada (p-value: {normal_p_value:.4f})")

# Test para la categoría "Paralelizado"
parallel_result, parallel_p_value = one_tailed_ttest(data_mint_parallel, data_ubuntu_parallel, alpha)
if parallel_result:
    print(f"En la categoría 'Paralelizado', Linux Mint tiene tiempos significativamente menores que Ubuntu (one-tailed, p-value: {parallel_p_value:.4f})")
else:
    print(f"En la categoría 'Paralelizado', NO se encontró una diferencia significativa en la dirección esperada (p-value: {parallel_p_value:.4f})")
```
